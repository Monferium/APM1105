# ITEM1
    Write a function named "ack" that evaluates the Ackermann function. Use your function to evaluate ack(2,4). What happens for larger values of m and n?

```python
#Create the function named "ack" that performs our Ackermman's Function
def ack(m, n):
    if m < 0 or n < 0:
        print("Infinite Recurrsion exist")
    if m == 0:
        return n + 1
    elif n == 0:
        return ack(m - 1, 1)
    else:
        return ack(m - 1, ack(m, n - 1))

#Calling the function to evaluate   
print(ack(2,4))
```

```python
#Observation A
m=int(input("input value for m: "))
n=int(input("input value for n: "))

print(ack(m,n))
```

```python
#Observation B
m=int(input("input value for m: "))
n=int(input("input value for n: "))

#Consider (3,-5)
print(ack(m,n))
```

```python
#Observation C
m=int(input("input value for m: "))
n=int(input("input value for n: "))

#Consider (7, 10)
print(ack(m,n))
```

### ANSWER
Observation A: When using small positive integers for m and n, many steps and recursive calls are necessary to obtain the final value.

Observation B: If either m or n is a negative integer, the function will perform infinite recursive calls since the conditions of m == 0 or n == 0 will never be met. This can cause the function to infinitely call itself.

Observation C: When using larger values for m and n, the number of steps and recursive calls grows exponentially before either m or n reaches zero. This tends to result in exceeding the maximum recursion depth when comparing multiple recursive function calls.

    Write a function called is_power that takes parameters a and b and returns True if a is a power of b. Note: you will have to think about a base case. (Note: It should work with float.)

```python
import math
def is_power(a, b):
    if a == int(a) and b == int(b):
        # base case: if a is 1, it is always a power of b
        if a == 1:
            return True
        
        #Since b cannot divides a; undefined
        if b == 0:
            return False
        
        # if a is divisible by b and a/b is a power of b, then a is a power of b
        if a % b == 0 and  is_power(a // b, b):
            return True

        # if none of the above conditions hold, a is not a power of b
        else:
            return False
    else:
        # Take the logarithm of a to the base b
        log = math.log(a, b)
        # Check if the result is an integer
        return log == int(log)
```
